```python
import re
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.naive_bayes import BernoulliNB,MultinomialNB,GaussianNB
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
from sklearn.pipeline import Pipeline
```

```python
sws=list(ENGLISH_STOP_WORDS)
sws.remove("not")
sws.remove("no")
for w in sws:
    if w.endswith("n't"):
        sws.remove(w)

def text_cleaning(doc):
    newdoc=re.sub(r"[^a-z0-9\s']",'',doc.lower())
    words=newdoc.split()
    newdoc=[]
    for word in words:
        if word not in sws:
            newdoc.append(word)
    return ' '.join(newdoc)
```

```python
df=pd.read_csv('c:/ml_datasets/Restaurant_Reviews.txt',sep='\t')
df.Review=df.Review.apply(text_cleaning)
df
```

```python
corpus=df.Review
target=df.Liked
```

```python
df1=pd.read_csv('c:/ml_datasets/Restaurant_Reviews.txt',sep='\t')
df2=pd.read_csv('c:/ml_datasets/more_reviews_restaurant.csv',sep='\t')
final_df=pd.concat([df1,df2],ignore_index=True)

corpus=final_df.Review
target=final_df.Liked

sws=list(ENGLISH_STOP_WORDS)
sws.remove("not")
sws.remove("no")
for w in sws:
    if w.endswith("n't"):
        sws.remove(w)

tv=TfidfVectorizer(lowercase=True,stop_words=sws)
pipe=Pipeline([('pre_vect',tv),('model',MultinomialNB())])
pipe.fit(corpus,target)
```
